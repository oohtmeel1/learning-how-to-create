```python


df_incidents.shape
```

```python
latitude = 39.1547
longitude = -77.2405
```

```python
maryland_map = folium.Map(location=[latitude, longitude], zoom_start=12)
maryland_map
```

```python
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.X, df_incidents.Y):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=20, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
maryland_map.add_child(incidents)
```
